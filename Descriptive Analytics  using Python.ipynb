```python
## Grouping and aggregating: To group records based on the column values and then apply aggregated operations such as mean, maximum etc. … example-Average sold price for each age category, group all records by age and then apply mean() on sold price columns
## Catagory1 has highetst average sold price
ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
## converting pd.series data structure to data frame by .rest_index()
soldprice_by_age = ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
soldprice_by_age
```

```python
## group by for multiple columns.. for age and palying role, converted in to df
soldprice_by_age_role = ipl_auction_df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
soldprice_by_age_role
```

```python
## Merge operation- Compare the average SOLDPRICE for different age catagories with different age and playing role catagories , merge the df soldprice_by_age and solprice_by_age role
```

```python
soldprice_by_age_role = ipl_auction_df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
```

```python
soldprice_by_age_role 
```

```python
##Joining DF: can combine feature of multiple df using pandas merge()
##Compare the average auction price for diﬀerent ages and playing roles
```

```python
soldprice_comparison = soldprice_by_age_role.merge( soldprice_by_age, on = 'AGE',  how = 'outer') 
```

```python
soldprice_comparison
```

```python
## because the column name SOLDPRICE is same in both data frame , it automatically renames them SOLD PRICE_x and  SOLD PRICE_y
```

```python
## compared the average SOLD PRICE for age catagories with different age and playing role catagories, merged the dataframes soldprice_by_age and soldprice_by_age_role
```

```python
soldprice_comparison.head(5) 
```

```python
##Renaming the columns: rename(), for renaming multiple columns simultaneously ,the method can take a dictionary as a parameter , where the keys should be existing column names and the values should be the new names to be assigned
soldprice_comparison.rename( columns = { 'SOLD PRICE_x': 'SOLD_PRICE_AGE_ROLE', 'SOLD PRICE_y': 'SOLD_PRICE_AGE' }, inplace = True  ) 
```

```python
soldprice_comparison.head(5)
```

```python
## Scatter plot example- Checking the influence b/w no of six hit and their soldprice
ipl_batsman_df = ipl_auction_df[ipl_auction_df['PLAYING ROLE'] == 'Batsman']
```

```python
## Pair plot: For identifying the relationship among many variables; Depict the relationship in a single diagram using pairpolt()
## Here identifying the relationship b/w SR-B', 'AVE', 'SIXERS', 'SOLD PRICE'which we think  may be the influential feaures to determine the soldpric
influential_features = ['SR-B', 'AVE', 'SIXERS', 'SOLD PRICE']
```

```python
sns.pairplot(ipl_auction_df[influential_features], size=2) 

#The pair plot technique allows us to visualize distributions of individual numerical features, like a matrix..
#as well as correlations/relationships between numerical features. 

## when we have many variable its not convenient to draw scatter plot for each pair of variable understanding 
#relationship we have to use pair plot, size
```

```python
## from all above correaltions, AVE & SIXERS seems to have highly correlated with SOLD PRICE compared to SR-B
```

```python
## Coorelation: Correlation used for measuring strength and direction of linear relationship between the continuous RV X & Y .It’s a statistical measure that indicates the extent to which two variables change together A Positive correlation means the variable increase or decrease together, A negative  correlation means if  one  variable increases the other decreases
ipl_auction_df[influential_features].corr()                  #Compute pairwise correlation of columns usig corr()
```

```python
## Heat map-Color map scale of correlation
##Setting annote attribute to true prints the correlation values in each box of the heat map and improves the readability of the heat map
## AVE & SIXER shows positive correlation while soldprice and SR-B are not so strongly correlated.
sns.heatmap(ipl_auction_df[influential_features].corr(), annot=True);  
```
